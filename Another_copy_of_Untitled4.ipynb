```python
import random

Codigos_livros = []
Livros_disponiveis = ["o pequeno principe", "vingadores"]
ID = ["pauloventura", "joabesilva"]
ID_ADM = ["pauloventura"]
livros_emprestados = []



def entrada():
 while True:

  print("""
 _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _
|                                      |
|   Bem vindo a Biblioteca Virtual     |
|                                      |
|_ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ |
        """)
  try:
       Entrada = str(input("Digite seu ID: \n"))
       print("")
       if Entrada in ID:
        print(f"Bem vindo {Entrada}\n")
        break
       else:
        print("Esse usuário não está cadastrado, cadastre novo usuário: ")
        Cadastro_ID = str(input("Digite o novo ID "))
        ID.append(Cadastro_ID)
        print("Cadastro realizado com sucesso!")

  except ValueError:
        print("Valor inválido, tente novamente!")
entrada()


def menu():
 while True:



    print("""
    --------------------
     Biblioteca Virtual
    --------------------
          """)
    print("[1] Pegar livro")
    print("[2] Devolver livro")
    print("[3] Cadastrar novo ID usuário")
    print("[4] Cadastrar novo livro")
    print("[5] Mostrar livros disponíveis")
    print("[6] Sair")
    print("[7] Administrador\n")


    try:

        Escolha = int(input("Escolha sua opção: "))

        if Escolha < 1 or Escolha > 7:
            print("Opção inválida!")
            continue

        if Escolha == 1:
            pegar = str(input("Qual livro você quer pegar? "))
            ID_Pegar = str(input("Qual seu ID? "))
            numero = random.randint(1, 100)
            if ID_Pegar in ID:
              print("ID confirmado com sucesso")
            else:
              print("Esse ID não existe!")
              continue
            if pegar in Livros_disponiveis:
              Livros_disponiveis.remove(pegar)
              livros_emprestados.append(pegar)
              Codigos_livros.append(numero)
              print(f"Livro emprestado com sucesso! E seu código é: {numero}\n")
              continue
            else:
              print("Esse livro não está na biblioteca!\n")
              continue

        elif Escolha == 2:
            Devolver = str(input("Qual livro você quer devolver? "))
            codigo = int(input("Qual o código do livro? "))
            if Devolver in livros_emprestados and codigo in Codigos_livros:
              print("Livro devolvido com sucesso!\n")
              Livros_disponiveis.append(Devolver)
              livros_emprestados.remove(Devolver)
              Codigos_livros.remove(codigo)
              continue
            else:
              print("Esse livro não consta em nosso banco de dados!")
              continue

        elif Escolha == 3:
            Cadastro_ID = str(input("Digite o novo ID "))
            ID.append(Cadastro_ID)
            print("Cadastro realizado com sucesso!\n")
            continue
        elif Escolha == 4:
           Cadastro_Livro = str(input("Digite o nome do livro: "))
           Codigos_livros.append(Cadastro_Livro)
           Livros_disponiveis.append(Cadastro_Livro)
           print("Novo livro cadastrado com sucesso!\n")
           continue
        elif Escolha == 5:
           print(Livros_disponiveis)
           continue
        elif Escolha == 6:
           print("Muito obrigdo pela visita! Até a próxima!\n")
           break

    except ValueError:
      print("Inválido, digite apenas números!")

    while True:

        if Escolha == 7:
           print("")
           entrada2 = str(input("Digite o ID de administrador: "))
           if entrada2 in ID_ADM:
            print("")
            print("Bem vindo PAULO")
           else:
            print("Essa entrada é apenas para administrador!")
            break
           print("")
           print("[1] Mostrar cadastros")
           print("[2] Mostrar códigos de livros")
           print("[3] descadastrar ID's")
           print("[4] Cadastrar ID ADM")
           print("[5] sair\n")
           Escolha2 = int(input("Escolha sua opção: "))
           if Escolha2 > 5 or Escolha2 < 1:
            print("Valor inválido! Tente novamente!")
           if Escolha2 == 1:
             print("")
             print(ID)
             print("")
             continue
           if Escolha2 == 2:
            print("")
            print(Codigos_livros)
            print("")
            continue
           if Escolha2 == 3:
            cancelar = str(input("Digite o ID que vc quer desvincular: "))
            if cancelar in ID:
              print("")
              print("ID removido com sucesso!")
              print("")
              ID.remove(cancelar)
              continue
            else:
              print("")
              print("ID não cadastrado! Tente novamente")
              print("")
              continue
           if Escolha2 == 4:
            Colocar_ADM = str(input("Bem vindo PAULO, Qual ID o senhor quer colocar como ADM?: "))
            ID_ADM.append(Colocar_ADM)
            print("ID cadastrado com sucesso!")
           if Escolha2 == 5:
              print("")
              print("Saindo do modo administrador")
              print("")
              break
menu()




```
